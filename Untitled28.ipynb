```
class fbiagent:
  def __init__(self, name, numberpushups, numbersitups):
    self.name = name
    self.numberpushups = numberpushups
    self.numbersitups = numbersitups
  
  def pointsinpushups(self):
    return self.numberpushups - 30
  def pointsinsitup(self):
    return self.numbersitups - 38
  def totalpoints(self):
    return self.numbersitups + self.numberpushups

nameofagent = input('name')
numberofpushupsofagent = int(input("how many pushups have you done"))
numberofsitupsofagent = int(input("how many situps have you done"))

chirstopher = fbiagent(nameofagent,numberofpushupsofagent,numberofsitupsofagent)

totalscore = input('do you want to check what is my total score')

if totalscore == 'yes':
  chirstopher.totalpoints()



```
